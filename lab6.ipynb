LinkedList class that we saw in the lecture (week 5). The code is in `lecture5.ipynb`.

```python
class Node:
    def __init__(self, d, n):
        self.data = d
        self.next = n

class LinkedList:
    def __init__(self):
        self.head = None
        self.length = 0

    def __str__(self):
        st = "--> "
        ptr = self.head
        while ptr != None:
            st = st + str(ptr.data)
            st = st+" -> "
            ptr = ptr.next
        return st+"None"
        
    def search(self, d):
        i = 0
        ptr = self.head
        while ptr != None:
            if ptr.data == d:
                return i
            ptr = ptr.next
            i += 1
        return -1
        
    def append(self, d):
        if self.head == None:      
            self.head = Node(d,None) 
        else:
            ptr = self.head
            while ptr.next != None:
                ptr = ptr.next
            ptr.next = Node(d,None)
        self.length += 1

    def insert(self, i, d):
        if self.head == None or i == 0:
            self.head = Node(d,self.head)
        else:
            ptr = self.head
            while i>1 and ptr.next != None:
                ptr = ptr.next
                i -= 1
            ptr.next = Node(d,ptr.next)
        self.length += 1

    def remove(self, i): # removes i-th element and returns it
        if self.head == None:
            return None
        if i == 0:
            val = self.head.data
            self.head = self.head.next
            self.length -= 1
            return val
        ptr = self.head
        while ptr.next != None:
            if i == 1:
                val = ptr.next.data
                ptr.next = ptr.next.next
                self.length -= 1
                return val                
            ptr = ptr.next
            i -= 1
    
    def removeVal(self, d):
        if self.head == None:
            return
        if self.head.data == d:
            self.head = self.head.next
            self.length -= 1
        else:
            ptr = self.head	
            while ptr.next != None:
                if ptr.next.data == d:
                    ptr.next = ptr.next.next
                    self.length -= 1
                    break
                ptr = ptr.next
    
    def sublist(self, i):
        ptr = self.head
        ls = LinkedList()
        ls.length = self.length
        while ptr != None and i>0:
            ptr = ptr.next
            i -= 1
            ls.length -= 1
        ls.head = ptr
        return ls
```

Add in `LinkedList` the following functions:

that appends to the linked list (the one represented by 
`self`) all elements of the array `A`. 
For example, if `ls` is `[2,3,4,3,5]` then `ls.appendAll([4,2,1])` changes `ls` to 
`[2,3,4,3,5,4,2,1]`.

Consider how efficient is your implementation -- can you do better?

b) `def nullify(self)`

that removes all elements from the list. I.e. `ls.nullify()` changes `ls` to `[]`.

c) `def clone(self)`

that creates a (new) copy of the list. For example, if `ls` is `[2,3,4,5]` then `ls.clone()` should return a new list `[2,3,4,5]`. The original list should not be changed.

d) `def merge(self, other)`

that merges the linked list with the linked list `other`. After merging, the list `other` should be empty and all its elements be part of `self`. For example, if `ls` is `[2,3,4,5]`
and `ls2` is `[42,17]`, then:

      ls.merge(ls2) 
      print(ls,",",ls.length,",",ls2,",",ls2.length)      
should print `--> 2 -> 3 -> 4 -> 5 -> 42 -> 17 ->| , 6 , -->| , 0`.

```python
def appendAll(self, A):
    for item in A:
        self.append(item)

def nullify(self):
    self.head = None
    self.length = 0

def clone(self):
    cloned_list = LinkedList()
    ptr = self.head
    while ptr != None:
        cloned_list.append(ptr.data)
        ptr = ptr.next
    return cloned_list

def merge(self, other):
    if not self.head:
        self.head = other.head
    else:
        ptr = self.head
        while ptr.next:
            ptr = ptr.next
        ptr.next = other.head
    self.length += other.length
    other.nullify()

LinkedList.appendAll = appendAll; LinkedList.nullify = nullify; LinkedList.clone = clone; LinkedList.merge = merge
```

```python
def insertRec(ptr, i, d):
    if ptr is None:
        return None
    if i == 0:
        return Node(d, ptr)
    ptr.next = insertRec(ptr.next, i-1, d)
    return ptr

def insert2(self, i, d):
    self.head = insertRec(self.head, i, d)

LinkedList.insert2 = insert2
```

Our implementation is based on a simple and naive idea: within each `ArrayListWithUndo` object we store a copy of the version of the object before the last operation was performed. Undoing the last operation then simply amounts to restoring that older copy. Note that for the miniproject you have to give a different implementation of `ArrayListWithUndo` which is based on a more subtle idea leading to a data structure that is more efficient. 

We define the new class as follows:

    class ArrayListWithUndo(ArrayList):
        def __init__(self):
            super().__init__()
            self.oldList = None
            
        def undo(self):
            if self.oldList == None: return
            temp = self.oldList.oldList
            self.inArray, self.count = self.oldList.inArray, self.oldList.count
            self.oldList = temp
            
The variable `oldList` will contain the copy of the array list before the last operation was performed. Initially, there is no older copy of the array list and `oldList` is set to `None`. The `undo` function simply restores the array and count of the `oldList` into `self`, and also updates `oldList` to `oldList.oldList`.
